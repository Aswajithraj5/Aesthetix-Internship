```python
#**nestedif**
x=int(input("enter a number= "))
if x>5:
  if x%2==0:
    print("x is greater than 5 and even")
  else:
    print("x is greater than 5 and odd")
elif x<5:
  if x%2==0:
    print("x is less than 5 and even")
  else:
    print("x is less than 5 and odd")
else:
  print("x is equal to 5")
```

```python
def mega(a, b, c):
  if a>=b and a>=c:
    return a
  elif b>=a and b>=c:
    return b
  else:
    return c
p,q,r=map(int,input("enter values=").split())
print("max =",mega(p,q,r))
```

```python

#Q1
def vowels(s):
  a=0
  v=["a","e","i","o","u"]
  for i in s:
    if i in v:
      a+=1
  return a
str=input("enter a string= ")
S=str.lower()
print("vowels=",vowels(S))
```

```python
#Q2
def prime(n):
  if n==2:
    return True
  for i in range(2,(int(n**0.5)+1)):
    if n%i==0:
      return False
    return True
num=int(input("enter a number= "))
print(prime(num))

```

```python
#Q3
def string(str,c):
  count=0
  for i in str:
    if i==c:
      count+=1
  return count

S=input("enter a string= ")
k=input("enter a character= ")
print(string(S,k))
```

```python
#Q4
def oddcount(n):
  count=0
  for i in str(n):
    if int(i)%2!=0:
      count+=1
  return count

n=int(input("enter a number= "))
print(oddcount(n))

```

```python
#Q5
def string(a):
  str=""
  for i in a:
    if i!=" ":
      str+=i
    else:
      pass
  return str

b=input("enter string=")
print(string(b))
```

```python
#Q6
def mini(a,b,c):
  if a<=b and a<=c:
    return a
  elif b<=a and b<=c:
    return b
  else:
    return c

p,q,r=map(int,input("enter three values=").split())
print("minimum=",mini(p,q,r))
```
